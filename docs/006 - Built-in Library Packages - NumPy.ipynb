## Built-in Library Packages in Python: NumPy

Today, we start looking into numerical computations.<br>
We will extend Pythons build-in commands by loading a so-called package that extends the capabilities of Python. <br>
In particular, we will use NumPy which is a **Numerical Python** package. To make use of some visualization aids, we also use the **matplotlib** library.

- NumPy is the fundamental package for scientific computing with Python 
- Documentation www.numpy.org
- It contains among other things:
- N-dimensional array object
- sophisticated (broadcasting) functions
- Interfacing with C/C++ and Fortran code
- useful linear algebra, Fourier transform, and random number capabilities
- Efficient multi-dimensional container of generic data.
- Arbitrary data-types can be defined
- Seamlessly and speedily integrate with a wide variety of databases.
- C core implementation

### 2. NumPy Array Object

- Arrays can be assigned to instances with the **array()** method
- The arguments can be a list for one-dimensional arrays or a nested list for **multidimensional** arrays
- A **dimension** of a array is refered to as axis
- We can use arrays to work with **vectors and matrixes**
- arrays are used to store **numerical values**

- The NumPty array object has four basis __attributes__ that are relevant for us
  - shape
  - size
  - ndim
  - dtype

```python
data = np.array ([[1, 2.], [3, 4], [5, 6]])

print('ndim attribute ', data.ndim)
print('shape attribute ',data.shape)
print('Dtype attribute ',data.dtype)
```

### 3. Defining a Function in NumPy

```python
data = np.array([2,4,3,6,5,4,6,6,7,4])

def mult(num):
    return num * np.array([2,4,3,6,5,4,6,6,7,4])

result = mult(2)
print("The multiplication result is", result)
```

```python
data = ([1,2,4,3,2,4,5,5,6,7,8,7,6,7,6,7,6,7,6,7,6,7,8])

def func_ave():
    return np.average(data)

print(func_ave())
```

```python
data = ["a", "5", "dua"]

def remover(data, index):
    return np.delete(data, index)
print(remover(data,1))
```

```python
import numpy as np
data = ([3,4,6,3,5,7,8,5,6])
identity = ([1,0,0], [0,1,0], [0,0,1])
def func():
    return np.reshape(data, (3,3)) * identity

print(func())
```

<div>
<h3>Simple plot</h3>
<p>In this section, we want to draw the cosine and sine functions on the same
plot. Starting from the default settings, we'll enrich the figure step by step
to make it nicer.</p>
<p>The first step is to get the data for the sine and cosine functions:</p>

```python
import numpy as np

X = np.linspace(0, 4*np.pi, 1000, endpoint=True)
C, S = np.cos(X), np.sin(X)
```

we need to load the library packages for ploting

```python
import matplotlib.pyplot as plt
```

now we ready to use it

```python
plt.plot(X,C)
plt.plot(X,S)
```

we need to put a legend

```python
plt.plot(X,C, label="Cos(x)")
plt.plot(X,S, label="Sin(x)")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

# Generate sample data for stock prices (replace this with your own data)
days = list(range(1, 31))
stock_prices = [100 + random.uniform(-500, 500) for _ in days]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(days, stock_prices, marker='o', linestyle='-')
plt.title('Stock Price Fluctuation Over Time')
plt.xlabel('Days')
plt.ylabel('Stock Price ($)')
plt.grid(True)

# Show the plot
plt.show()
```
